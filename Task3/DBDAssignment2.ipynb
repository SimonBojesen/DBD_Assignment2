```python
def toBinary(string):
    return "".join([format(ord(char),'#010b')[2:] for char in string])
```

```python
str1 = 'beebs beepps!!!!!  their eerie ears hear pears'
bits1 = toBinary(str1)
bitstr1 = str(bits1)
print('Bit length of Original ASCII bits: ' + str(len(bitstr1)))

wrongbitstr2 = '1111110101111111011111101011101110000010010010010010110111100010100011011110101001100111011010011000110001010011110111010011000'
print('Bit length of Wrong Bit str (missing the a character): ' + str(len(wrongbitstr2)))

correctbitstr2 = '1011010110111101111101101011100110011010010010010010011111010010000110101000111010100010101011110110010001101111100001100100011111000110010001101'
print('Bit length of Correct Bit str: ' + str(len(correctbitstr2)))
```
